```python
def carga(*args) -> list:
    """
    #carga(str)->list
    Recibe un archivo de tipo csv y devuelve dos listas, una con sublistas que representan cada fila,
    y otra con el encabezado.
    """
    with open(*args, "r", encoding="UTF-8") as contenido:
        lector = csv.reader(contenido)
        lista_retorno = []
        encabezado = next(lector)
        for fila in lector:
          lista_retorno.append(fila) #agrega cada fila (representada con una lista), a la lista vacía.

    return lista_retorno, encabezado

#Prueba

prueba, encabezado = carga(archivo) #ejecuto la función con el archivo, y guardo las dos listas en las variables prueba y encabezado respectivamente.
print(prueba)
print(encabezado)
```

```python
class EstacionMeteorologica:
    """
    Esta clase representa una estación meteorológica,
    contiene los datos relevantes al análisis.
    """
    def __init__(self, nombre_estacion, temps_prom_mens, temp_maxima, temp_minima,
                 hum_promedio, hum_maxima, hum_minima, vel_promedio, vel_maxima, vel_minima):
        self.nombre_estacion = nombre_estacion
        self.temps_prom_mens = temps_prom_mens
        self.temp_maxima = temp_maxima
        self.temp_minima = temp_minima
        self.hum_promedio = hum_promedio
        self.hum_maxima = hum_maxima
        self.hum_minima = hum_minima
        self.vel_promedio = vel_promedio
        self.vel_maxima = vel_maxima
        self.vel_minima = vel_minima

    def __str__(self):
        return f"El presente objeto representa a la estación meteorológica {self.nombre_estacion}. Cada estación provee información sobre {self.temps_prom_mens}, {self.temps_prom_mens}, {self.temp_maxima}, {self.temp_minima}, {self.hum_promedio}, {self.hum_maxima}, {self.hum_minima}, {self.vel_promedio}, {self.vel_maxima}, {self.vel_minima}."
```
